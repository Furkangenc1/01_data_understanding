```python


## Cell 4 – Load datasets
app_event = pd.read_csv("app_events.csv")
complaint = pd.read_csv("complaints.csv")
customer = pd.read_csv("customers.csv")
data_dictionary = pd.read_csv("data_dictionary.csv")
drivers = pd.read_csv("drivers.csv")
hubs = pd.read_csv("hubs.csv")
incidents = pd.read_csv("incidents.csv")
orders = pd.read_csv("orders.csv")
vehicle = pd.read_csv("vehicles.csv")

print("All datasets loaded successfully")
```

Cell 6 – Dataset shape (rows and columns)

```python
datasets = {
    "app_event": app_event,
    "complaint": complaint,
    "customer": customer,
    "drivers": drivers,
    "hubs": hubs,
    "incidents": incidents,
    "orders": orders,
    "vehicle": vehicle
}

for name, df in datasets.items():
    print(f"{name} shape:", df.shape)
```

Cell 7 – Column names

```python
for name, df in datasets.items():
    print(f"\n{name} columns:")
    print(df.columns.tolist())
```

Cell 8 – Data types

```python
for name, df in datasets.items():
    print(f"\n{name} data types:")
    print(df.dtypes)
```

Cell 9 – Missing values

```python
missing_values = {}

for name, df in datasets.items():
    missing = df.isnull().sum()
    missing_values[name] = missing
    print(f"\nMissing values in {name}")
    print(missing[missing > 0])
```

Cell 10 – Duplicate records

```python
for name, df in datasets.items():
    duplicates = df.duplicated().sum()
    print(f"{name} duplicate rows: {duplicates}")
```

Cell 11 – Summary statistics (numerical columns)

```python
for name, df in datasets.items():
    print(f"\nSummary statistics for {name}")
    display(df.describe())
```

Cell 11 – Summary statistics (numerical columns)

```python
for name, df in datasets.items():
    print(f"\nSummary statistics for {name}")
    display(df.describe())
```

Cell 12 – Unique values for categorical columns

```python
for name, df in datasets.items():
    print(f"\n{name} categorical values")

    categorical_cols = df.select_dtypes(include='object').columns

    for col in categorical_cols:
        unique_vals = df[col].nunique()
        print(f"{col}: {unique_vals} unique values")
```

Cell 13 – Inspect zone inconsistencies

```python
print("Zones in customers:")
print(customer['home_zone'].unique())

print("\nDriver base zones:")
print(drivers['base_zone'].unique())

print("\nHub zones:")
print(hubs['zone'].unique())

print("\nOrder pickup zones:")
print(orders['pickup_zone'].unique())

print("\nVehicle assigned zones:")
print(vehicle['assigned_zone'].unique())
```

Cell 14 – Date columns check

```python
date_columns = {
    "app_event": "event_timestamp",
    "complaint": "created_at",
    "customer": "signup_date",
    "incidents": "reported_at",
    "orders": "order_created_at",
    "vehicle": "commission_date"
}

for dataset, col in date_columns.items():
    print(f"\nChecking {dataset} date column")

    df = datasets.get(dataset)

    if col in df.columns:
        print(df[col].head())
```
